## 1_D 時刻表示

Sの秒数からmの分数への変換が難しい。

演算子としては、//で商の整数を取得すればもっと楽。

```python
S= int(input())
h=int(S/3600)
m=int((S-h*3600)/60)
s=int(S%60)
print(f"{h}:{m}:{s}")
```

## 2_B 範囲

ワンライナーの条件式の書き方が参考になった

```python
a, b, c= map(int, input().split() )
print ("%s" %("Yes"if a<b<c else"No"))
```

## 2_C 3 つの数の整列

読み込んだ文字列をsortで評価して、再度スペースを含む出力にすることがポイント

```python
x = map(int, input().split() )
print (*sorted(x))
```

## 2_D 長方形の中の円

条件を思いつくかがポイント。
人によって書き方が大きく分かれる。

```python
W,H,x,y,r = map(int, input().split())

if 0 <= x-r <= W and 0 <= x+r <= W and 0 <= y-r <= H and 0 <= y+r <= H:
  print ("Yes")
else:
  print ("No")
```

## 3_B テストケースの出力

関数を定義する人は少ない印象。

```python
def main(i,x):
    return (f"Case {i}: {x}")
    
i=1
while i <=10000:
    x=int(input())
    if x ==0:
        break
    else:
        print (main(i,x))
        i+=1
```

## 3_C 2 つの数の交換

while Trueを利用してみた。

sortの出力がポイント

```python
while True:
    a,b = map(int, input().split() )
    if a == 0 and b == 0:
        break
    print (*sorted([a, b]))
```

## 3_D 約数の数
